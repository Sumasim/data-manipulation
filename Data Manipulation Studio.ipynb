## Clean Your Data

```python
# Clean your data here!
for col in data_Baltimore.columns:
    pct_missing = np.mean(data_Baltimore[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
    
data_Baltimore.head()

#modifying Repack column values

data_Baltimore["Repack"] = data_Baltimore["Repack"].replace({"N":False , "E":True})

data_Baltimore.head(3)

#shape

data_Baltimore.shape

data_Baltimore = data_Baltimore[['Type', 'Grade', 'Origin District', 'Environment', 'Quality', 'Condition', 'Appearance', 'Storage', 'Crop', 'Trans Mode', 'Sub Variety', 'Color']]



data_Baltimore.shape

data_Baltimore.head()
```

```python
for col in data_Boston.columns:
    pct_missing = np.mean(data_Boston[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
    
data_Boston["Repack"] = data_Boston["Repack"].replace({"N":False , "E":True})

data_Boston.head(3)

data_Boston.shape

data_Boston = data_Boston.drop(['Type', 'Sub Variety', 'Grade', 'Origin District', 'Environment', 'Quality', 'Condition', 'Appearance', 'Storage', 'Crop', 'Trans Mode'], axis=1)

data_Boston.head()

data_Boston.shape
```

```python
# New York data cleanup

for col in data_NewYork.columns:
    pct_missing = np.mean(data_NewYork[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
    
data_NewYork["Repack"] = data_NewYork["Repack"].replace({"N":False , "E":True})

data_NewYork.head(3)

data_NewYork.shape

data_NewYork = data_NewYork.drop(['Type', 'Sub Variety', 'Grade', 'Origin District', 'Color', 'Environment', 'Quality', 'Condition', 'Appearance', 'Storage', 'Crop', 'Trans Mode'], axis=1)

data_NewYork.head()

data_NewYork.shape
```

## Combine Your Data

Now that you have four clean sets of data, combine all four into one dataframe that represents the entire Northeast region.

```python
# Combine the four dataframes into one!
data_Northeast = pd.concat([data_Baltimore, data_Boston, data_NewYork, data_Philadelphia], keys = ["Baltimore", "Boston", "NewYork", "Philadelphia"])
data_Northeast
```

## Answer Some Questions

Use `groupby()` and `agg()` to answer the following two questions:

1. What is the mean low and high prices for each type of unit of sale in the Northeast region? In the last lesson, we learned that a unit of sale could be something like a bin or individually. 
2. What is the average number of pumpkins for each variety that came into terminal markets for the year by region? Pumpkin varieties include Howden and Fairytale pumpkins.

```python
# Put your code here to find the mean low and high prices in the Northeast region for each type of unit of sale.
data_Northeast1 = data_Northeast.drop(['Repack'], axis=1)

data_Northeast1['Unit of Sale'] = data_Northeast1['Unit of Sale'].fillna('UNKNOWN')
data_Northeast1

data_Northeast1_grp = data_Northeast1.groupby('Unit of Sale')

data_Northeast1_grp

data_Northeast1_grp.first()

agg_func_list = ['mean', 'min', 'max']

data_Northeast1_grp.agg(agg_func_list) 
```

```python
# Put your code here to find the average number of pumpkins coming into terminal markets of each variety.
data_Northeast2 = data_Northeast.drop(['Repack'], axis=1)

data_Northeast2['Variety'] = data_Northeast1['Variety'].fillna('UNKNOWN')
data_Northeast2

data_Northeast2_grp = data_Northeast2.groupby('Variety')

data_Northeast2_grp

data_Northeast2_grp.first()

data_pumpkin = data_Northeast2['Variety'].value_counts()
data_pumpkin

# df = data_pumpkin.rename_axis('Variety').reset_index(name='counts')
# print (df)

df = data_pumpkin.rename_axis('Variety').to_frame('counts')
print (df)

agg_func_list = ['mean']

df.agg(agg_func_list)
```
